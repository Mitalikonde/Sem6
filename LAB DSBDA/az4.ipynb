```python
#Import python libraries

import pandas as pd
import numpy as np
```

```python
#Load the dataset and store it in a dataframe.

df = pd.read_csv("BostonHousing.csv")
df.info()
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#Removing null value by replacing it with mean

df['rm'] = df['rm'].fillna(df['rm'].mean())
```

```python
df.isnull().sum()
```

```python
#Here we seperate the first 13 columns in X and the last column to be predicted as Y.

X = df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
Y = df['medv']
```

```python
#Splitting the data into training and testing data using train_test_split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3, random_state = 42)
```

```python
#Importing and loading the Linear Regression model on the data.

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
print(y_pred)
```

```python
#Root Mean Squared Error:

from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mead squared Error is:")
print(rmse)
```

```python
#Training Accuracy:

print("Training accuracy is:")
lr.score(X_train, y_train)
```

```python
#Testing Accuracy:

print("Testing accuracy is:")
lr.score(X_test, y_test)
```
